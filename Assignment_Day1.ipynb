3. Write a function named season and accept one parameter named weather.
Weather variable should accept an input. Using conditional statements write a code to check if the weather is cold, then it should print Winter. If the weather is sunny, then it should print Summer Else print wrong input

```
def season(weather):
  if(weather == "cold"):
    print("Winter")
  elif(weather == "sunny"):
    print("Summer")
  else:
    print("Wrong Input")

weather = input("Enter the Weather: ")
season(weather)
```

5. Write a code for Countdown timer using loops and time library for 10 seconds

```
import time
i = 1
while(i <=10):
  print(i,end = '\t')
  time.sleep(1)
  i = i + 1
```
